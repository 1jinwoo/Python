# Course Objectives
* Develop fluency in the Python Language
    - pretty much the entire language  will be covered, except for some obscure topics
    - get you more or less as good as you are in the language you already know
- Learn about common tasks and issues
* Understand some of the Python "pros and cons"
* Sample the huge space of Python libraries
* Do a project in Python - something of interest to you


# Prerequisites 
- Course directory: "Intended only for those who are already fluent in at least one programming language."
- Ideally:
    - you've written at least some small programs
    - have a basic understanding of concepts like:
      - functions and argument passing
      - call stacks
      - iteration
      - recursion
      - basic data structures like:
          - lists
          - strings 
          - hash/map/dictionary tables
      - pointers/references

- You do not have to be a Computer Science major, or a programming wizard
- You do not need prior experience in Object Oriented Programing 
- I will make some comparisons with Java/C++, but ok if you don't know those languages
- I will "pop the hood" a few times to look at Python implementation and performance issues.
    - This is optional material and I don't spend much time on it. 
    - If too technical for you, don't worry about it.
- If you don't have something like above background, still possible to do the class, but NOT recommended

# Resources
* We will be using the Anaconda
5.2 Python 3.6 distribution. It has hundreds of [packages](https://docs.anaconda.com/anaconda/packages/py3.6_osx-64) built in.
    - goto [Anaconda](https://www.anaconda.com/download/) and scroll down a bit
    - make sure you download the 3.6 graphical installer, not the 2.7 one
    - Anaconda is nice in that it does not "mess your machine up". it makes one directory in your home directory, and adds one entry to your PATH variable. straightforward to uninstall if you don't want to keep it
    - you may already have a python program on your machine, but it will be the wrong one - you must install anaconda
    - the installer will present a sequence of screens
        - when you see "Change Install Location..." in the bottom right, click on that. then click on "Install for me only". continue
    - after installation, type:
    - python -V
    - you should see something like:
    - Python 3.6.5 :: Anaconda, Inc.
    - will use 'jupyter notebooks' as our primary Python environment(included in Anaconda)
        - to setup jupyter:
        - enter a simple password by typing in a shell window
        - jupyter-notebook password
        - jupyter will prompt you for the password now and then
        - if you forget your password, just set it again
        - to run the jupyter server, 'cd' to the directory that has your notebooks and type
        - jupyter-notebook
    - if you like IDE's I recommend 'spyder' (included in Anaconda)
